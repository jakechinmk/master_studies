# Question 4

```r
car = c('Audi A8',
        'BMW 328Xi',
        'Cadillac CTS',
        'Chryster 300',
        'Ford Focus',
        'Hyundai Elantra',
        'Jeep Grand Cherokee',
        'Pontiac G6',
        'Toyota Camry',
        'Volkswagen Jetta'
       )

size = c('Large',
         'Compact',
         'Midsize',
         'Large',
         'Compact',
         'Midsize',
         'Midsize',
         'Compact',
         'Midsize',
         'Compact'
        )

cylinders = c(12,
              6,
              6,
              8,
              4,
              4,
              6,
              6,
              4,
              5
             )

citympg = c(13,
            17,
            16,
            13,
            24,
            25,
            17,
            15,
            21,
            21
           )

highwaympg = c(19,
               25,
               25,
               18,
               33,
               33,
               26,
               22,
               31,
               29
              )

fuel = c('Premium',
         'Premium',
         'Regular',
         'Premium',
         'Regular',
         'Regular',
         'Diesel',
         'Regular',
         'Regular',
         'Regular'
        )

df1 = data.frame(Car=car, 
                Size=size, 
                Cylinders=cylinders, 
                'City MPG'=citympg,
                'Highway MPG'=highwaympg,
                Fuel=fuel
               )
df1
```

```r
mean(citympg)
mean(highwaympg)-mean(citympg)
plot(citympg,highwaympg)
cor(citympg,highwaympg)
```

```r
library(ggplot2)
library(reshape2)
```

```r
manufacturer<-c("General Motors","Ford","DaimlerChrysler","Toyota")
year2004<-c(8.9,7.8,4.1,7.8)
year2005<-c(9.0,7.7,4.2,8.3)
year2006<-c(8.9,7.8,4.3,9.1)
year2007<-c(8.8,7.9,4.6,9.6)
df2<-data.frame(manufacturer,year2004,year2005,year2006,year2007)
summary(df2[,2:5])
var(df2[,2:5])
#mean close to median, variance similar--normal#
df2
data<-melt(df2,id.vars="manufacturer")
data
attach(data)
ggplot(data,aes(x=as.numeric(variable),y=value,color=manufacturer))+geom_line()+labs(x="Year")
barplot(year2007,names.arg = manufacturer)
```
