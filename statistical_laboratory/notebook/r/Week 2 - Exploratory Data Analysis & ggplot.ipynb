```r
g = ggplot(mpg, aes(cty)) + geom_density(aes(fill=factor(cyl))) #fill to a factor variable
g + geom_density(aes(fill=factor(cyl))) +
 labs(title="Density plot",
 subtitle="City Mileage Grouped by Number of cylinders",
 caption="Source: mpg",
 x="City Mileage",
 fill="No of Cylinders")
```

```r
g = ggplot(mpg, aes(class, cty))
g + geom_boxplot(varwidth=T, fill="orange") +
labs(title="Box plot",
 subtitle="City Mileage grouped by Class of vehicle",
 caption="Source: mpg",
 x="Class of Vehicle",
 y="City Mileage")
```

```r
#basic scatter plot
ggplot(mpg, aes(x = cty, y = hwy)) +
 geom_point()
#basic scatter plot with labels
ggplot(mpg, aes(x = cty, y = hwy)) +
 geom_point() +
 labs(title="Scatter plot",
 subtitle="City vs Highway Mileage",
 caption="Source: mpg",
 x="City Mileage",
 y="Highway Mileage")
#scatterplot with groups
ggplot(mpg, aes(x = cty, y = hwy)) +
 geom_point(aes(colour=factor(cyl)))
#transformation using log function
ggplot(mpg, aes(x = log(cty), y = hwy)) +
 geom_point(aes(colour=factor(cyl)))
#scatter plot with fitted values usign linear regression
ggplot(mpg, aes(x =cty, y = hwy)) +
 geom_point(aes(colour=factor(cyl))) + stat_smooth(method = "lm")
#scatter plot with loess smoothing function
gg=ggplot(mpg, aes(x = cty, y = hwy)) +
geom_point(aes(colour=factor(cyl))) + geom_smooth(method="loess",
se=F)
```

```r
#basic scatter
# load package and data
library(ggplot2)
library(ggExtra)
data(mpg, package="ggplot2")
# Scatterplot
theme_set(theme_bw()) # pre-set the bw theme.
g <- ggplot(mpg, aes(cty, hwy)) + geom_point () +
geom_smooth(method="lm", se=F)
#adding histogram/boxplot/density on the scatter plot
ggMarginal(g, type = "histogram", fill="transparent")
#ggMarginal(g, type = "boxplot", fill="transparent")
# ggMarginal(g, type = "density", fill="transparent")
```

```r
# load package and data
library(ggplot2)
data(mpg, package="ggplot2")
# selecting 4 manufacturers only
mpg_select <- mpg[mpg$manufacturer %in%
 c("audi", "ford", "honda", "hyundai"), ]
# Scatterplot
g <- ggplot(mpg_select, aes(displ, cty)) +
 labs(subtitle="mpg: Displacement in liters vs City Mileage",
 title="Bubble chart")
#adding the data by 4 selected manufacturer
g + geom_jitter(aes(colour=manufacturer))
#adding the data by 4 selected manufacturer and bubble size for highway mileage
g + geom_jitter(aes(colour=manufacturer, size=hwy))
#adding the fitted linear model line to different manufacturers on the same plot
g + geom_jitter(aes(colour=manufacturer, size=hwy)) +
 geom_smooth(aes(col=manufacturer), method="lm", se=F)
```

```r
install.packages("ggcorrplot")
library(ggcorrplot) #library(ggplot2)
# Correlation matrix
data(mtcars)
corr <- round(cor(mtcars), 1) #round to 1 decimal places
# Plot
ggcorrplot(corr,
 type = "lower", lab = TRUE,
 method="circle",
 colors = c("tomato2", "white", "springgreen3"),
 title="Correlogram of mtcars")
```
